```python
import datetime
def convert_time(time):
    try:
        return datetime.datetime.strptime(str(time), "%Y-%m-%d")
    except:
        try:
            return datetime.datetime.strptime(str(time), "%Y-%m-%d %H:%M:%S")
        except:
            try:
                return datetime.datetime.strptime(str(time), "%Y%m%d%H%M%S")
            except:
                try:
                    return datetime.datetime.strptime(str(time), "%Y%m%d")
                except:
                    try:
                        return datetime.datetime.strptime(str(time), "%d%m%Y%H%M%S")
                    except:
                        print("Wrong date format: %s" % time)
                        return None
now_time = datetime.datetime.now() 
data['person_birthday'] = data['IDNO'].str[6:14].apply(convert_time)
data['age'] = data['person_birthday'].apply(lambda x:(now_time-x).days/365).astype(int)
data
```
